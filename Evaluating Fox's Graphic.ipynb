```python
import pandas as pd

from dateutil.parser import parse
from dateutil.relativedelta import relativedelta

from scipy import stats

from IPython.core.display import HTML
```

```python
sp500 = pd.read_csv('data/^GSPC.csv')
```

```python
def from_str(date):
    return parse(date)


def to_str(date):
    return date.strftime('%Y-%m-%d')


def one_week_later(date):
    return to_str(from_str(date) + relativedelta(weeks=1))
```

```python
BEFORE = -1
AFTER = 1
def nearest_stock_date(stocks, date, direction):
    original_date = date
    pdate = from_str(date)
    i = 0
    while date not in sp500['Date'].values:
        pdate += relativedelta(days=direction)
        date = to_str(pdate)
        i += 1
        if i > 5:
            print('Failed to find valid date after 5 tries for original date:', original_date)
            break
    return date
```

```python
def get_events_effect_on_sp500(events, stocks):
    df = []
    for event, (date, other_person) in events.items():
        date_week_later = nearest_stock_date(stocks, one_week_later(date), AFTER)
        date = nearest_stock_date(stocks, date, BEFORE)
        change = stocks[stocks['Date'].isin((date, date_week_later))]['Adj Close'].pct_change().iloc[-1]
        df.append({
            'Event': event,
            'Date': date,
            'pct_change': change * 100,
            'Percent Change': f'{change * 100:0.1f}%',
            'Other Person': other_person,
        })

    return pd.DataFrame(df)
```

```python
print(f"Average change: {effect_from_all_events['pct_change'].dropna().mean():0.1f}%")
_, pval = stats.ttest_1samp(effect_from_all_events['pct_change'].dropna(), 0)
print(f'p = {pval:0.4f}')
```

## Normalize by Preceding Week's Change

```python
def get_events_effect_on_sp500_normalized_by_preceding_week(events, stocks):
    df = []
    for event, (date, other_person) in events.items():
        week_end = nearest_stock_date(stocks, one_week_later(date), AFTER)
        week_start = nearest_stock_date(stocks, date, BEFORE)
        week_change = stocks[stocks['Date'].isin((week_end, week_start))]['Adj Close'].pct_change().iloc[-1]
        
        preceding_week_end = week_start
        preceding_week_start = nearest_stock_date(stocks, to_str(from_str(preceding_week_end) + relativedelta(weeks=-1)), BEFORE)
        preceding_week_change = stocks[stocks['Date'].isin((preceding_week_start, preceding_week_end))]['Adj Close'].pct_change().iloc[-1]
        
        df.append({
            'Event': event,
            'Date': date,
            'pct_change': week_change,
            'preceeding_pct_change': preceding_week_change,
            'Other Person': other_person,
        })

    df = pd.DataFrame(df)
    df['norm_pct_change'] = df['pct_change'] - df['preceeding_pct_change']
    return df
```

```python
effect_from_all_events_norm = get_events_effect_on_sp500_normalized_by_preceding_week(all_events, sp500)
display(HTML(effect_from_all_events_norm.dropna().sort_values('norm_pct_change').to_html()))
```

```python
print(f"Average change: {effect_from_all_events_norm['norm_pct_change'].dropna().mean()}%")
_, pval = stats.ttest_1samp(effect_from_all_events_norm['norm_pct_change'].dropna(), 0)
print(f'p = {pval:0.4f}')
```

So after taking into account how the stock market had been doing, it appears the S&P 500 still is not significantly affected by the events and definitely not in the positive direction.
